```python
# Import dependencies
import requests
import json
from pprint import pprint
```

```python
# url variable
url = 'https://earthquake.usgs.gov/earthquakes/feed/v1.0/summary/significant_month.geojson'
```

```python
# Grab data and jsonify it
earthquakes = requests.get(url)
earthquakesJSON = earthquakes.json()
```

```python
# Pretty print the dat to inspect it
pprint(earthquakesJSON)
```
